(45 points). Write a python program to conduct N-gram analysis based on the dataset in your assignment two:

(1) Count the frequency of all the N-grams (N=3).

(2) Calculate the probabilities for all the bigrams in the dataset by using the fomular count(w2 w1) / count(w2). For example, count(really like) / count(really) = 1 / 3 = 0.33.

(3) Extract all the **noun phrases** and calculate the relative probabilities of each review in terms of other reviews (abstracts, or tweets) by using the fomular frequency (noun phrase) / max frequency (noun phrase) on the whole dataset.
```python
import pandas as pd
import regex as re
from nltk.corpus import stopwords
import nltk
from textblob import TextBlob
nltk.download('stopwords')
nltk.download('brown')
nltk.download('punkt')
stopwords = nltk.corpus.stopwords.words('english')
```

```python
#reading csv from Github
df=pd.read_csv('https://raw.githubusercontent.com/ravilladivyaUNT/Assignment-three/main/Reviews.csv',sep=',')
print(df.head(5).to_string())
```

```python
#checking for null and duplicate values
print(df.any().isna())
print('Note there are no null values in a dataset\n')
print(df[df.duplicated()])
print('Note there are no duplicates in a dataset')
```

```python
def ngrams(text, n):
    text = text.lower()
    text = re.sub(r'[^a-zA-Z0-9\s]', ' ', text)
    words = [word for word in text.split(" ") if word not in stopwords and word!='']

    ngrams = zip(*[words[i:] for i in range(n)])
    return list(ngrams)

df.Review_of_the_movie=df.Review_of_the_movie.astype(str)
trigramValues= [word for word in list(df['Review_of_the_movie'].apply(lambda x:ngrams(x,3))) if len(word)!=0]
bigramValues= [word for word in list(df['Review_of_the_movie'].apply(lambda x:ngrams(x,2))) if len(word)!=0]
unigramValues= [word for word in list(df['Review_of_the_movie'].apply(lambda x:ngrams(x,1))) if len(word)!=0]

print("Trigrams:",trigramValues)
```

```python
trigramsCount=0
for listValue in trigramValues:
  trigramsCount+=len(listValue)
print("\nFrequency of Tri-grams",trigramsCount)
```

```python
bigrams=[]
for Tuples in bigramValues:
  for Tuple in Tuples:
    bigrams.append(Tuple)
print("Bigrams:",bigrams)
```

```python
unigrams=[]
for Tuples in unigramValues:
  for Tuple in Tuples:
    for value in Tuple:
     unigrams.append(value)
print("Unigrams",unigrams)
```

```python
probabilityCount={}
for value in bigrams:
  probabilityCount[value]=round(bigrams.count(value)/unigrams.count(value[0]),2)
print(probabilityCount)
```

```python
def getNouns(txt):
  blob = TextBlob(txt)
  return blob.noun_phrases
df['Nouns']=df['Review_of_the_movie'].apply(lambda x:getNouns(x))
print(df[['Nouns','Review_of_the_movie']].head(20).to_string(index=False))
```

# **Question 2: Undersand TF-IDF and Document representation**

(20 points). Starting from the documents (all the reviews, or abstracts, or tweets) collected for assignment two, write a python program: 

(1) To build the **documents-terms weights (tf*idf) matrix bold text**.

(2) To rank the documents with respect to query (design a query by yourself, for example, "An Outstanding movie with a haunting performance and best character development") by using **cosine similarity**.

```python
import pandas as pd
import string
from bs4 import BeautifulSoup
from sklearn.feature_extraction.text import TfidfVectorizer
from nltk.tokenize.toktok import ToktokTokenizer
import math
from collections import Counter
nltk.download('wordnet')
```

```python
#reading csv from github
df=pd.read_csv('https://raw.githubusercontent.com/ravilladivyaUNT/Assignment-three/main/Reviews.csv',sep=',')
df.dropna(inplace=True)
print(df.head(5).to_string(index=False))
```

```python
def removePunctuation(text):
    punctuationfree="".join([i for i in text if i not in string.punctuation and i.isdigit()==False])
    return punctuationfree

df.Review_of_the_movie=df.Review_of_the_movie.astype(str)
df['Review_of_the_movie']=df['Review_of_the_movie'].apply(lambda x:removePunctuation(x))
print(df['Review_of_the_movie'].head(5).to_string(index=False))
```

```python
#changing the text to lowercase
df['Review_of_the_movie']= df['Review_of_the_movie'].apply(lambda x: x.lower())
print(df['Review_of_the_movie'].head(5).to_string(index=False))
```

```python
def removeStopWords(txt):
  return " ".join([word for word in txt.split(' ') if word not in stopwords])

df['Review_of_the_movie']=df['Review_of_the_movie'].apply(lambda x:removeStopWords(x))
print(df['Review_of_the_movie'].head(5).to_string(index=False))
```

```python
df['Review_of_the_movie'].to_csv("data.csv",index=True)
```

```python
corpus = list(df['Review_of_the_movie'])
vectorizer = TfidfVectorizer()
matrix = vectorizer.fit_transform(corpus)
print(matrix)
```

```python
WORD = re.compile(r"\w+")

def cosine(v1, v2):    
    intersection = set(v1.keys()) & set(v2.keys())
    numerator = sum([v1[x] * v2[x] for x in intersection])

    sum1 = sum([v1[x] ** 2 for x in list(v1.keys())])
    sum2 = sum([v2[x] ** 2 for x in list(v2.keys())])
    denominator = math.sqrt(sum1) * math.sqrt(sum2)

    if not denominator:
        return 0.0
    else:
        return float(numerator) / denominator

def vector(text):
    words = WORD.findall(text)
    return Counter(words)

query = 'marvel fully enjoyable movie'
result=pd.DataFrame(columns=['Query','Document','Similarity'])

v1 = vector(query)
for review in df['Review_of_the_movie']:
  v2 = vector(review)
  result.loc[len(result.index)] = [query, review, cosine(v1, v2)] 

result.sort_values(by='Similarity',ascending=False,inplace=True)
print(result.head(20).to_string(index=False))
```
